```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
"""
First, we need to load the images.
Therefore, let's determine the size of each image we load.
"""

def load(img_folder: str, start: int, stop: int, step: int):
    img_sizes = range(start, stop, step)

    n_rows = int(np.sqrt(len(img_sizes)))
    n_cols = int(np.sqrt(len(img_sizes)))

    fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(8, 8))

    for i, img_size in enumerate(img_sizes):
        sample_class_dir =  os.listdir(img_folder)[2]
        sample_image =  os.listdir(os.path.join(img_folder, sample_class_dir))[0]
        image_path = os.path.join(img_folder, sample_class_dir, sample_image)
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        image = cv2.resize(image, (img_size, img_size), interpolation=cv2.INTER_AREA)
        axes[i % n_rows, i // n_cols].imshow(image, cmap='gray', vmin=0, vmax=255)
        axes[i % n_rows, i // n_cols].set_title('Size of: (' + str(img_size) + ',' + str(img_size) + ')')
        plt.tight_layout()

    return fig
```

```python
fig = load('./outputs/', 50, 500, 50)
```

After displaying the image sizes in the range of 50 to 450,
It seems that for 50 by 50 size images the blur is significant while for 250 by 250 size
images or more, the blur is negligible, and the various elements can be clearly identified.
